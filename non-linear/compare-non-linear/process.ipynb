```python
import numpy as np
import matplotlib.pyplot as plt
with open('/home/jinzhuo/ww_dataset/DECOMPOSITION/non-linear/compare-non-linear/rmse.txt', 'r') as f:  
    data = f.readlines()  

lst = []
for i in range(len(data)):
    if i%7 !=0:
        s = data[i]
        s = s.replace('\n', '')
        number = float(s)
        lst.append(number)

sub_lst = [lst[i:i+6] for i in range(0, len(lst), 6)]
sub_lst.reverse()
```

```python
labels = ['200','300','400','500','600','700','800','900','1000']
x = np.arange(len(labels))
sub_lst = np.array(sub_lst)
```

```python

plt.figure(figsize=(20,8))
width = 0.2
plt.bar(2*x-2.5*width-0.02,sub_lst[x,0],width,label ='cell0-DNN',color = '#934B43')
plt.bar(2*x-1.5*width-0.02,sub_lst[x,1],width,label ='cell1-DNN',color = '#D76364')
plt.bar(2*x-0.5*width-0.02,sub_lst[x,2],width,label ='cell2-DNN',color = '#EF7A6D')
plt.bar(2*x+0.5*width+0.02,sub_lst[x,3],width,label ='cell0-Linear',color = '#9394E7')
plt.bar(2*x+1.5*width+0.02,sub_lst[x,4],width,label ='cell1-Linear',color = '#5F97D2')
plt.bar(2*x+2.5*width+0.02,sub_lst[x,5],width,label ='cell2-Linear',color = '#9DC3E7')
plt.xticks(2*x,labels = labels,fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Trainset(Reference) single cell number',fontsize = 18)
plt.ylabel('RMSE',fontsize = 18)
plt.legend(loc = 'upper right',fontsize = 20)
plt.savefig("/home/jinzhuo/ww_dataset/DECOMPOSITION/non-linear/compare-non-linear/rmse.svg",format = 'svg',dpi = 300)
```
